```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Electric_data = pd.read_csv("/drive/My Drive/Colab Notebooks/Hamoye related/UCI_Electrical_grid_stability.csv")
Electric_grid_stability_dataframe = Electric_data.copy()
Electric_grid_stability_dataframe.head()
```

```python
Electric_grid_stability_dataframe.info()
```

```python
Electric_grid_stability_dataframe.describe()
```

```python
Electric_grid_stability_dataframe.drop("stabf", axis=1)
```

```python
Electric_grid_stability_dataframe.stabf = Electric_grid_stability_dataframe.stab
ones = Electric_grid_stability_dataframe.query("stab <= 0")
for x in Electric_grid_stability_dataframe.stabf:
  if x <= 0:
    Electric_grid_stability_dataframe.stabf.replace(x, 1, inplace=True)
  else:
    Electric_grid_stability_dataframe.stabf.replace(x, 0, inplace=True)
                                                    
Electric_grid_stability_dataframe.head()                                                    
```

```python
features = Electric_grid_stability_dataframe.drop(columns=["stab", "stabf"])
features
```

```python
label = Electric_grid_stability_dataframe["stabf"]
label
```

```python
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
features_normalized = ss.fit_transform(features, label)
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain, xtest, ytrain, ytest = tts(features_normalized, label, test_size=.2, random_state=1)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1)
x_train, y_train = smote.fit_resample(xtrain, ytrain)
```

```python
from sklearn.ensemble import RandomForestClassifier

x_train_rfc = x_train.copy()
y_train_rfc =  y_train.copy()
x_test_rfc = xtest.copy()
y_test_rfc = ytest.copy()

rfc = RandomForestClassifier()
rfc.fit(x_train_rfc, y_train_rfc)
```

```python
rfc_predict = rfc.predict(x_test_rfc)
```

```python
# importing metrics for evaluation
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test_rfc, rfc_predict))
print(classification_report(y_test_rfc, rfc_predict))
```

```python
from sklearn.ensemble import ExtraTreesClassifier

x_train_etcr = x_train.copy()
y_train_etcr =  y_train.copy()
x_test_etcr = xtest.copy()
y_test_etcr = ytest.copy()

etcr = ExtraTreesClassifier()
etcr.fit(x_train_etcr, y_train_etcr)
```

```python
print(etcr.feature_importances_)
```

```python
etcr_predict = etcr.predict(x_test_etcr)
```

```python
print(confusion_matrix(y_test_etcr, etcr_predict))
print(classification_report(y_test_etcr, etcr_predict))
```

```python
from xgboost import XGBModel

x_train_xgbm = x_train.copy()
y_train_xgbm =  y_train.copy()
x_test_xgmb = xtest.copy()
y_test_xgbm = ytest.copy()

xgbm = XGBModel()
xgbm.fit(x_train_xgbm, y_train_xgbm)
```

```python
xgbm_predict = xgbm.predict(x_test_xgmb)
```

```python
from sklearn.utils.fixes import sklearn
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("mean absolute error: ", mean_absolute_error(y_test_xgbm, xgbm_predict))
msError = mean_squared_error(y_test_xgbm, xgbm_predict)
print("mean square error: ", msError)
print("Root mean squared error: ", np.sqrt(msError))
```
